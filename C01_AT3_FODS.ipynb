```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()

df = pd.read_csv("education_dataset.csv")

print("========== EDUCATION ANALYTICS DATASET ==========\n")
print(df)

print("\n1. Number of Rows and Columns")
print(df.shape)

print("\n2. Data Types")
print(df.dtypes)

print("\n3. Missing Values")
print(df.isnull().sum())

print("\nDuplicate Values")
print(df.duplicated().sum())

print("\n4. Statistical Summary")
print(df.describe())

print("\nMedian")
print(df.median(numeric_only=True))

plt.figure(figsize=(6,4))
df["Result"].value_counts().plot(kind="bar")
plt.title("Pass and Fail Count")
plt.xlabel("Result")
plt.ylabel("Number of Students")
plt.show()

plt.figure(figsize=(6,4))
plt.hist(df["Attendance"], bins=5)
plt.title("Attendance Distribution")
plt.xlabel("Attendance")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df["Study_Hours"], df["Internal_Marks"])
plt.title("Study Hours vs Internal Marks")
plt.xlabel("Study Hours")
plt.ylabel("Internal Marks")
plt.show()

plt.figure(figsize=(6,4))
plt.boxplot(df["Attendance"])
plt.title("Attendance Box Plot")
plt.ylabel("Attendance")
plt.show()

print("\n========== OBSERVATIONS ==========")
print("1. Students with higher attendance mostly passed the examination.")
print("2. Students who studied more hours obtained higher internal marks.")
print("3. Students with low attendance generally failed.")
print("4. Internal marks increase as study hours increase.")
print("5. Good attendance and regular study improve student performance.")

print("\n========== CONCLUSION ==========")
print("Attendance and study hours have a positive impact on student performance. Students with higher attendance and better internal marks are more likely to pass. The EDA indicates that maintaining regular attendance and consistent study habits can improve academic results.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()

df = pd.read_csv("banking_loan_dataset.csv")

print("========== BANKING LOAN APPROVAL DATASET ==========\n")
print(df)

print("\n1. Number of Rows and Columns")
print(df.shape)

print("\n2. Data Types")
print(df.dtypes)

print("\n3. Missing Values")
print(df.isnull().sum())

print("\nDuplicate Values")
print(df.duplicated().sum())

print("\n4. Statistical Summary")
print(df.describe())

print("\nMedian")
print(df.median(numeric_only=True))

plt.figure(figsize=(6,4))
df["Loan_Status"].value_counts().plot(kind="bar")
plt.title("Loan Approval Status")
plt.xlabel("Loan Status")
plt.ylabel("Number of Customers")
plt.show()

plt.figure(figsize=(6,4))
plt.hist(df["Income"], bins=5)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df["Credit_Score"], df["Loan_Amount"])
plt.title("Credit Score vs Loan Amount")
plt.xlabel("Credit Score")
plt.ylabel("Loan Amount")
plt.show()

plt.figure(figsize=(6,4))
plt.boxplot(df["Credit_Score"])
plt.title("Credit Score Box Plot")
plt.ylabel("Credit Score")
plt.show()

print("\n========== OBSERVATIONS ==========")
print("1. Customers with higher credit scores are more likely to have their loans approved.")
print("2. Higher-income customers generally receive higher loan amounts.")
print("3. Most rejected applications belong to customers with lower incomes and credit scores.")
print("4. Salaried employees have a higher loan approval rate than self-employed and unemployed applicants.")
print("5. Credit score plays a significant role in loan approval decisions.")

print("\n========== CONCLUSION ==========")
print("The EDA shows that income, credit score, and employment type are the major factors influencing loan approval. Customers with stable employment, higher incomes, and good credit scores are more likely to receive loan approval.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()

df = pd.read_csv("agriculture_crop_dataset.csv")

print("========== AGRICULTURE CROP YIELD DATASET ==========\n")
print(df)

print("\n1. Number of Rows and Columns")
print(df.shape)

print("\n2. Data Types")
print(df.dtypes)

print("\n3. Missing Values")
print(df.isnull().sum())

print("\nDuplicate Values")
print(df.duplicated().sum())

print("\n4. Statistical Summary")
print(df.describe())

print("\nMedian")
print(df.median(numeric_only=True))

plt.figure(figsize=(6,4))
df["Soil_Type"].value_counts().plot(kind="bar")
plt.title("Number of Farms by Soil Type")
plt.xlabel("Soil Type")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(6,4))
plt.hist(df["Crop_Yield_kg"], bins=5)
plt.title("Crop Yield Distribution")
plt.xlabel("Crop Yield (kg)")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df["Rainfall_mm"], df["Crop_Yield_kg"])
plt.title("Rainfall vs Crop Yield")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Crop Yield (kg)")
plt.show()

plt.figure(figsize=(6,4))
plt.boxplot(df["Fertilizer_kg"])
plt.title("Fertilizer Usage Box Plot")
plt.ylabel("Fertilizer (kg)")
plt.show()

print("\n========== OBSERVATIONS ==========")
print("1. Farms with higher rainfall generally produce higher crop yields.")
print("2. Loamy soil records the highest crop yield among all soil types.")
print("3. Higher fertilizer usage is associated with better crop production.")
print("4. Extremely high temperatures are linked to lower crop yields.")
print("5. Rainfall, fertilizer usage, and soil type are key factors influencing crop production.")

print("\n========== CONCLUSION ==========")
print("The EDA shows that crop yield is strongly influenced by rainfall, fertilizer usage, and soil type. Farms with adequate rainfall, proper fertilizer application, and loamy soil achieve better crop production. Maintaining these factors can help improve agricultural productivity.")
```
